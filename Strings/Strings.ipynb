```python
import string
def str_to_int(s):
    r = 0
    if s[0] == '-':
        t = s[s[0]=='-':]
    else:
        t = s
    for k,v in enumerate(t):
        r += string.digits.index(v) * 10**(len(t) - 1 - k)
    if s[0] == '-':
        return r * -1
    return r
```

```python
str_to_int('909090')
```

```python
str_to_int('9697')
```

```python
str_to_int('590')
```

```python
str_to_int('-590')
```

```python
import functools, string
def s_to_i(s):
    return functools.reduce(
        lambda r, c: r * 10 + string.digits.index(c), s[s[0]=='-':], 0
    ) * (-1 if s[0]=='-' else 1)
```

```python
def int_to_str(x):
    s = []
    is_neg = False
    if x < 0:
        x, is_neg = -x, True
    while x:
        s.append(chr(ord('0') + x % 10))
        x //= 10
    if is_neg:
        return ('-') + ''.join(reversed(s))
    return ''.join(reversed(s))
```
